### VÃ©rification des valeurs abberantes

```python
# Identification des donnÃ©es aberrantes - IQR
# Fonction pour dÃ©tecter les outliers en utilisant l'IQR
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)]
```

```python
detect_outliers_iqr(df, "V7").sort_values("V7")
```

## Traitement des Outliers

```python
colonnes = df.select_dtypes(include="float64").columns

for var in colonnes:
    Q1 = df[var].quantile(0.25)
    Q3 = df[var].quantile(0.75)
    IQR = Q3 - Q1

    borne_inf = Q1 - 1.5 * IQR
    borne_sup = Q3 + 1.5 * IQR

    df.loc[df[var] < borne_inf, var] = borne_inf
    df.loc[df[var] > borne_sup, var] = borne_sup
```

## Tests statistiques
#### Test de Mann Whitney

```python
from scipy.stats import mannwhitneyu
def mann_whitney_test(df,variable,target="Class"):
    groupe_0= df[df[target]==0][variable].dropna()
    groupe_1= df[df[target]==1][variable].dropna()
    stat,p_value= mannwhitneyu(groupe_0,groupe_1, alternative="two-sided")

    return stat, p_value
```

```python
variable= downsampled.select_dtypes(include= ["int64", "float64"]).columns
variables= [v for v in variable if v!= "Class"]

result= []
for var in variables:
    stat,p=mann_whitney_test(df, var)
    result.append({
        "variable": var,
        "stat_MW":stat,
        "p_value": p
    })

result_MW= pd.DataFrame(result).sort_values("p_value")
print(result_MW)
```

#### Calcul du VIF

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def compute_vif(df,parametre):
    X=downsampled[parametre].dropna()
    vif_data= pd.DataFrame()
    vif_data["variable"]=X.columns
    vif_data["VIF"]= [variance_inflation_factor(X.values, i)
                      for i in range(X.shape[1])
                      ]
    return vif_data.sort_values("VIF", ascending=False)
```

## division des donnÃ©es

```python
X= downsampled.drop(["Class", "V13","V22","V15","V11"], axis=1)
y= downsampled["Class"]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.30,random_state= 111)
```

```python
from sklearn.model_selection import cross_val_score, KFold
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn import model_selection
from xgboost import XGBClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import classification_report,explained_variance_score, roc_curve, roc_auc_score
```

## Construction du ModÃ¨le

```python
models = []
models.append(('Log', LogisticRegression()))
models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('KNN', KNeighborsClassifier()))
models.append(('CART', DecisionTreeClassifier()))
models.append(('RF', RandomForestClassifier()))
models.append(('NB', GaussianNB()))
models.append(('SVMr', SVC(kernel= 'rbf')))
models.append(('SVMl', SVC(kernel= 'linear')))
models.append(('GBC', GradientBoostingClassifier()))
models.append(('XGB', XGBClassifier()))

results = []
names = []
for name,model in models:
    kfold = model_selection.KFold(n_splits=10, shuffle=True, random_state= 0)
    cv_result = cross_val_score(model,X_train,y_train, cv = kfold, scoring ='f1')
    names.append(name)
    results.append(cv_result)
for i in range(len(names)):
    print(names[i], ':',results[i].mean())
```

```python
model= LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_test_pred = model.predict(X_test)
y_train_pred = model.predict(X_train)
```

## Matrice de confusion

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_test_pred)

%matplotlib inline
label = ["0","1"]
plt.figure(figsize = (8, 5))
sns.heatmap(cm, annot=True)
plt.xlabel("Predicted")
plt.ylabel("RÃ©alitÃ©")
plt.title('La matrice de confusion')

plt.show()
```

```python
print('Classification: \n>>>>', classification_report(y_test, y_test_pred))
```

```python
# Calcul des probabilitÃ©s pour chaque classe
train_probs = model.predict_proba(X_train)[:, 1]
test_probs = model.predict_proba(X_test)[:, 1]

# Calcul des courbes ROC
train_fpr, train_tpr, _ = roc_curve(y_train, train_probs)
test_fpr, test_tpr, _ = roc_curve(y_test, test_probs)

# Calcul des AUC
train_auc = roc_auc_score(y_train, train_probs)
test_auc = roc_auc_score(y_test, test_probs)

# TracÃ© des courbes ROC
plt.figure(figsize=(10, 6))
plt.plot(train_fpr, train_tpr, label=f'Train ROC (AUC = {train_auc:.2f})')
plt.plot(test_fpr, test_tpr, label=f'Test ROC (AUC = {test_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```
